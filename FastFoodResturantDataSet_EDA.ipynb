### 3.3 Preprocessing

- Dealing with missing values<br/>
    - Dropping rows which has missing entries of __Websites.__
    - Dropping __Country__ column as it does not affect EDA and it is US only.
    

```python
fastfoodresturant_data.dropna(inplace = True)
fastfoodresturant_data.info()
```

```python
fastfoodresturant_data.drop(['country'],axis=1,inplace = True)
```

- Reanalyzing the dataframe<br/>
    - __Country__ column is removed now

```python
fastfoodresturant_data.info()
```

- Checking the __unique fast food resturant names__ present in __name__ column of Fast Food Resturant dataframe
- It is observed that __many unique resturant names__ are __having bit different names__. 

```python
fastfoodresturant_data.name.unique()
```

- Cleaning of __similar name resturant__ and __grouping them as one name__

```python
fastfoodresturant_data.name.replace(to_replace = 'Mc.*',value = 'Mc Donalds',regex = True,inplace = True )
fastfoodresturant_data.name.replace(to_replace = 'KFC.*',value = 'KFC',regex = True,inplace = True )
fastfoodresturant_data.name.replace(to_replace = 'Domino.*',value = 'Dominos Pizza',regex = True,inplace = True )
fastfoodresturant_data.name.replace(to_replace = 'Burgerville.*',value = 'Burgerville',regex = True,inplace = True )

fastfoodresturant_data.name.replace(to_replace = 'Sonic.*',value = 'Sonic Drive-In',regex = True,inplace = True )
fastfoodresturant_data.name.replace(to_replace = 'SONIC Drive In',value = 'Sonic Drive-In',regex = True,inplace = True )
fastfoodresturant_data.name.replace(to_replace = 'SUBWAY.*',value = 'Subway',regex = True,inplace = True )
fastfoodresturant_data.name.replace(to_replace = 'A&w.*',value = 'A & W',regex = True,inplace = True )
```

- __Sorting of whole dataframe__ on basis of __name column__

```python
fastfoodresturant_data.sort_values('name',inplace = True)
```

- Again analyzing the name column

```python
fastfoodresturant_data.name.any
```

### 4. Questions

### 4.1 Which restaurant has the highest number of branches in US?

- Using GroupBy Function & Bar Charts

```python
topTenResturant = fastfoodresturant_data['name'].value_counts().to_frame()
topTen = topTenResturant.groupby(['name']).head(5)
print (topTen)
```

```python
fastfoodresturant_data['name'].value_counts().head(10).plot.bar()
```

- ##__Observation Results__ - From above analysis, we found that __Mc Donalds__ fast food chain resturant has the highest number of branches in US followed by __Burger King__ & __Taco Bell__.

### 4.2 Which region in US is densely populated with fast food chain restaurants?

- Using Scatter Plot

```python
fastfoodresturant_data['province'] = fastfoodresturant_data['province'].astype('category').cat.codes
ax = fastfoodresturant_data.plot(kind="scatter", x="longitude", y="latitude", c="province", cmap=plt.get_cmap("jet"),
        label='Fast Food Restaurants', title='Fast Food Restaurants Location Map', figsize=(20,8))
ax.set_xlabel("longitude")
```

- ##__Observation Results__ - From above analysis, we found that __California(CA)__ province has the highest number of fast food chain restaurants in US followed by __Texas(TX)__ and __Ohio(OH)__.

### 4.3 Which city in US has highest number of fast food chain restaurants?

```python
fastfoodresturant_data['city'].value_counts().head(10).plot.bar()
```

- ##__Observation Results__ - From above analysis, we found that __Cincinnati__ city has the highest number of fast food chain restaurants in US followed by __Las Vegas__ and __Houston__.

## 5. Conclusions

- From the above analysis, we can draw the following conclusion:-
    1. Resturants which majorly involves __Burger__ as main dish like __Mc Donalds & Burger King__ is doing the most flourishing business in US. So, any new chain fast food resturant which involves __Burger__ as __main dish__ is __more likely to get success__ in US.
    2. Provinces which are near to __sea/seaport__ like __California, Texas and Ohio__ are densely populated with the people who like Fast Food. So opening of __Fast Food Chain Resturants in these provinces is more likely to get success__.
    3. Cities which have __nearby port__ or are __fun capital__ like __Cincinnati, Las Vegas & Houston__ attracts tourists & business a lot. So __opening up of resturants in these cities__ are more likely to get good business.
